# Répartition de la taille des graphes

```python
from matplotlib import pyplot as plt
from batch_GAT import faire_datasets_grph
import numpy as np

(ds,) = faire_datasets_grph(train=True, dev=False, test=False)
```

## Nombre de graphes dans le dataset

```python
print("Il y a %d graphes dans le jeu d’entrainement"%len(ds))
```

```python
repartition = np.array([d.msk1.shape[0] for d in ds])
```

```python
def afficher_histo(repart, bins=None):
    fig, axes = plt.subplots()
    axes.hist(repart, bins=bins)
    fig.show()
```

```python
afficher_histo(repartition, bins=200)
```

Une grande partie des graphes a moins de 2500 sommets

```python
repart2 = repartition[repartition <= 2500]
print("Il y a %d graphes de moins de 2500 sommets"%len(repart2))
```

```python
afficher_histo(repart2, bins=50)
```

```python
repart3 = repartition[repartition <= 1500]
print("Il y a %d graphes de moins de 1500 sommets"%len(repart3))
ratio = len(repart3)/len(repartition)
print("%f %% des graphes ont moins de 1500 sommets"%(100*ratio))
```

## 97,08% des graphes ont moins de 1500 sommets

```python
afficher_histo(repart3, bins=50)
```
